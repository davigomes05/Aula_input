```python
nome = " MARIA EDUARDA ALMEIDA "
email = " MARIA.ALMEIDA@EMPRESA.COM "

#1- Remover espaços extras
nome = nome.strip()
email = email.strip()

#2- Colocar tudo em minúsculo
nome = nome.lower()
email = email.lower()

#3- Substistituir email
email = email.replace("empresa.com", "gmail.com")

#4- Quantidade de caracteres
quantidade = len(nome)

#5- ultimo/primeiro
ultima_letra = nome[-1]
primeira_letra = nome[0]
print("Nome tratado:", nome)
print("Email tratado:", email)
print("Quantidade de caracteres:", quantidade)
print("Última letra:", ultima_letra)
print("Primeira letra:", primeira_letra)
```
